```python
from skimage import transform,data
import matplotlib.pyplot as plt
img = data.camera()
# dst=transform.resize(img, (80, 60))
dst=transform.rescale(img,scale=2,order=0,anti_aliasing=True,preserve_range=True)
plt.figure('resize')
 
plt.subplot(121)
plt.title('before resize')
plt.imshow(img,plt.cm.gray)
 
plt.subplot(122)
plt.title('before resize')
plt.imshow(dst,plt.cm.gray)
 
plt.show()
```

```python
import numpy as np
img=np.eye(10)#,dtype=np.bool)
img[2,3]=2
# dst=transform.resize(img, (20, 20), order=0, preserve_range=True)
dst=transform.rescale(img,scale=2,order=0,anti_aliasing=True,preserve_range=True)

plt.subplot(121)
plt.title('before resize')
plt.imshow(img,plt.cm.gray)
 
plt.subplot(122)
plt.title('before resize')
plt.imshow(dst,plt.cm.gray)
 
plt.show()
np.unique(dst)
```

```python
import numpy as np
from numpy import random
img=random.randint(0,255,(10,10,10))

# img[2,3]=2
# dst=transform.resize(img, (20, 20, 20), order=0)
dst=transform.rescale(img,scale=2,order=0,anti_aliasing=True,preserve_range=True)

n = 3
plt.figure(dpi=200)
plt.subplot(141)
plt.title('before resize')
plt.imshow(img[:,:,n],plt.cm.gray)
 
plt.subplot(142)
plt.title('before resize')
plt.imshow(dst[:,:,n*2-1],plt.cm.gray)

plt.subplot(143)
plt.title('after resize')
plt.imshow(dst[:,:,n*2],plt.cm.gray)

plt.subplot(144)
plt.title('before resize')
plt.imshow(dst[:,:,n*2+1],plt.cm.gray)
 
plt.show()
np.unique(dst)
```
